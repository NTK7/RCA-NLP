```python
import pandas as pd
```

```python
df = pd.read_csv('./data/data.csv')
df.head()
```

```python
df.shape
```

```python
df.dropna(subset=['review_text'], inplace=True)
```

```python
df.shape
```

```python
df.tail()
```

```python
df.drop(['google_id', 'place_id', 'location_link', 'reviews_link', 'reviews_per_score', 'review_id', 'author_link', 'author_title', 'author_id', 'name', 'author_image', 'review_img_url', 'owner_answer', 'owner_answer_timestamp', 'owner_answer_timestamp_datetime_utc', 'review_link', 'rating', 'review_timestamp', 'review_datetime_utc', 'review_likes', 'reviews_id'], axis=1, inplace=True)
df.head()
```

```python
# getting the minimum and maximum values of the rating column
print("min", df['review_rating'].min())
print("max", df['review_rating'].max())
```

```python
df['label'] = df['review_rating'].apply(lambda x: 1 if x > 3 else 0)
df.drop(['review_rating'], axis=1, inplace=True)
df.head()
```

```python
# get total count of positive and negative reviews
df['label'].value_counts()
```

```python
df.shape
```

```python
# get reviews with label 0
df[df['label'] == 0]
```

```python
# get reviews with label 0
df[df['label'] == 1]
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

```python
df.to_csv('./data/data_cleaned.csv', index=False)
```
